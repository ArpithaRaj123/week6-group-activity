```python
#importing python libreries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
#importing and printing the dataset
data=pd.read_excel('Dataset_w6.xlsx')
```

# 1. Find out how many countries are there in the dataset.

```python
data['country'].count()
```

# 2. Examine the data by using visualizations.

```python
#plotting first visualization usinng scatterplot
plt.figure(figsize=(6,6))
plt.title('Region and Population wrt GDP per capita',size="18")
sns.scatterplot(x="Population in thousands (2017)",y="Region",hue="GDP per capita (current US$)",data=data)
```

```python
#plotting second visualization usig scatterplot
plt.figure(figsize=(6,6),)
plt.title('Region and Population wrt International trade:Imports',size="18")
sns.scatterplot(x="Population in thousands (2017)",y="Region",hue="International trade: Imports (million US$)",data=data)
```

```python
#plotting 3rd visualization using pairplot
pp=sns.pairplot(data,hue='Region')
pp.fig.suptitle("PAIRPLOT VISUALIZATION",size="15")
plt.show()
```

# 3. Find out how many countries with respect to the region are available in the dataset

```python
# displaying country and regions
data[['country','Region']]
```

```python
data.groupby('Region')['country'].count()
```

# 4. a)On average which region has the highest GDP per capita (current Us$)

```python
#display average GDP per capita of each region
avg_gdp=data.groupby('Region')['GDP per capita (current US$)'].mean()
avg_gdp
```

```python
#region having highest GDP per capita
avg_gdp.nlargest(1)
```

# 4. b)On average which region has the highest international trade: Imports (million US$)

```python
#display average internationaltrade:Imports of each region
avg_trade=data.groupby('Region')['International trade: Imports (million US$)'].mean()
avg_trade
```

```python
#region having highest international trade:imports
avg_trade.nlargest(1)
```

# 5. Find out which region has more consistent with respect to GDP percapita (current US$)

```python
#display standard deviation GDP per capita of each region
sd=data.groupby('Region')['GDP per capita (current US$)'].std()
sd
```

```python
#as standard deviation decreases,consistency increases
sd.nsmallest(1)
```
